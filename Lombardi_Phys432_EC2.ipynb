```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

$$
\text{Solutions for } \vec{E} \times \vec{B} \text{ drift in Griffith's}
$$
$$
y(t) = C_1 \cos(\omega t) + C_2 \sin(\omega t) + \frac{E}{B}t + C_3,
$$

$$
z(t) = C_2 \cos(\omega t) - C_1 \sin(\omega t) + C_4.
$$


```python
# constants
q_e = -1.6e-19  # Electron charge (C)
m_e = 9.11e-31  # Electron mass (kg)
E = 1e3         # Electric field magnitude (V/m) 
B = 1           # Magnetic field magnitude (T)
omega = q_e * B / m_e  # Cyclotron frequency
v_drift = E / B        # Drift velocity
dt = 1e-9             # Time step (s)
t_max = 1e-6          # Total simulation time (s)
steps = int(t_max / dt)
```

```python
# Initialize vector for velocity and position
y = np.zeros(steps)
z = np.zeros(steps)
t = np.linspace(0, t_max, steps)
```

```python
# Initial conditions to determine constants
y[0] = 0
z[0] = 0
vy = 1e5 
vz = 0    
```

```python
# Get numerical solution
for i in range(1, steps):
    t_i = t[i]
    y[i] = vy / omega * np.sin(omega * t_i) + v_drift * t_i
    z[i] = -vy / omega * np.cos(omega * t_i) + vy / omega
```

```python
# Plot 
plt.figure(figsize=(10, 6))
plt.plot(y, z)
plt.xlabel("y (m)")
plt.ylabel("z (m)")
plt.title("E X B Trajectory")
plt.show()
```
