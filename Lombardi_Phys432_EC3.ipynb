```python
import numpy as np
import matplotlib.pyplot as plt

# Dipole moment
p = 1.0

# Electric field components in Cartesian
def E_field(x, y):
    r = np.sqrt(x**2 + y**2) + 1e-9  
    theta = np.arctan2(y, x)
    E_r = (2 * np.cos(theta)) / r**3
    E_theta = (np.sin(theta)) / r**3
    E_x = E_r * np.cos(theta) - E_theta * np.sin(theta)
    E_y = E_r * np.sin(theta) + E_theta * np.cos(theta)
    return E_x, E_y

# Meshgrid Points
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

# Compute field 
E_x, E_y = E_field(X, Y)

# Stream plot
plt.figure(figsize=(8, 8))
plt.streamplot(X, Y, E_x, E_y,
               density=0.9, linewidth=1, cmap='autumn')  # density adjusted for 25 lines
plt.title('Electric Field of a Perfect Dipole')
plt.plot(2, 0, 'bo')  # Mark the point (2, 0)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.show()
```

### The line integral around any closed loop in the field of a dipole is zero. This is because the electric field of a static charge distribution is conservative
